# Example run of `nfield`

Here's a quick runthrough of how to run the class. First import stuff...

```python
import sys
path = '/Users/Rob/work/nfield'
sys.path.append(path + '/source/') 
from nfield import nfield
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Then setup initial conditions and parameters like field masses through the Hubble and potential gradient. The number of dimensions is automatically set in `nfield` using the initial ensemble input (here `phi0`) so make sure this is consistent with your defined functions.

```python
phi0 = np.random.normal(0.001,0.1,size=(2,1000))
deltaN = 0.01
Nefolds = 10.0
m1, m2 = 2.0, 3.0
Hend = 1.0

def Vprime(phi,N):
    Vprime = np.zeros_like(phi)
    Vprime[0,:] = m1*phi[0,:]
    Vprime[1,:] = m2*phi[1,:]
    return Vprime

def H(phi,N):
    return Hend

# Example of how to define a first passage problem
def firstpass_func(prevphi,phi,N):
    passing = np.zeros((2,phi.shape[0],phi.shape[1]),dtype=bool)
    phi_thresh = 0.013
    another_phi_thresh = 0.016
    passing[0,(prevphi<phi_thresh)&(phi>=phi_thresh)] = True
    passing[1,(prevphi<another_phi_thresh)&(phi>=another_phi_thresh)] = True
    return passing
```

To run `nfield`, just write, e.g.,...

```python
nf = nfield(Vprime,H)
nf.ens._snapshot_rate = 5
nf.ens.firstpass_func = firstpass_func
nf.setup(phi0,deltaN)
nf.run(Nefolds)
```

Which runs the ensemble system and automatically computes and outputs the covariance matrix between fields at each timestep (this can be switched off by setting `nf.ens._store_covmat_isTrue=False` if desired). The data dictionary `nf.ens.dataDict` now contains all of the output information requested. For example, the covariance matrix at different times... 

```python
plt.plot(
    nf.ens.dataDict['CovMatrix']['t'],
    np.asarray(nf.ens.dataDict['CovMatrix']['x'])[:,0,0]
)
plt.plot(
    nf.ens.dataDict['CovMatrix']['t'],
    np.asarray(nf.ens.dataDict['CovMatrix']['x'])[:,1,1]
)
axes = plt.gca()
plt.show()
```

The full set of samples at various time snapshots...

```python
plt.hist(nf.ens.dataDict['Snapshots']['x'][0][0])
plt.hist(nf.ens.dataDict['Snapshots']['x'][100][0])
plt.show()
```
